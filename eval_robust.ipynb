```python
import pandas as pd
import os
from eval_v2_robust import eval_robust

robust_folder = '/home/haitong/PycharmProjects/lvrep-rl-cloned/log/Pendulum-v1_sigma_0.0_rew_scale_1.0/rfsac_robust_True_nystrom_False_rf_num_512_learn_rf_False'
nonrobust_folder = '/home/haitong/PycharmProjects/lvrep-rl-cloned/log/Pendulum-v1_sigma_0.0_rew_scale_1.0/rfsac_robust_False_nystrom_False_rf_num_512_learn_rf_False'

dfs_robust_multi = []
dfs_nonrobust_multi = []
for dir in os.listdir(robust_folder):
    print(dir)
    if dir.startswith('skip'):
        continue
    exp_path = os.path.join(robust_folder, dir)
    dfs_robust_multi.append(eval_robust(exp_path))
for dir in os.listdir(nonrobust_folder):
    exp_path = os.path.join(nonrobust_folder, dir)
    dfs_nonrobust_multi.append(eval_robust(exp_path))
df_robust = pd.concat(dfs_robust_multi, ignore_index=True)
df_nonrobust = pd.concat(dfs_nonrobust_multi, ignore_index=True)
```
